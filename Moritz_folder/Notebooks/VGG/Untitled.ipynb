```python
def meancalc(Hessian_diag_x):
    vals = []
    for entry in Hessian_diag_x:
        mean = torch.mean(entry)
        vals.append(mean.item())
    return vals

layermeans = {
            'zero_weight': [],
            'zero_bias': [],
            'one_weight': [],
            'one_bias': [],
            'three_weight': [],
            'three_bias': [],
            'six_weight': [],
            'six_bias': [],
            'eleven_weight': [],
            'eleven_bias': [],
            'thirteen_weight': [],
            'thirteen_bias': [],
            'sixteen_weight': [],
            'sixteen_bias': [],
            'eighteen_weight': [],
            'eighteen_bias': [],
            'twentythree_weight': [],
            'twentythree_bias': [],
            'twentysix_weight': [],
            'twentysix_bias': [],
            'twentyeight_weight': [],
            'twentyeight_bias': []
        }

for i in [10, 11, 12, 13, 14, 15]:
    state = torch.load('./trained_weights/ckpt_seed{}.pth'.format(s), map_location=torch.device('cpu'))
    Hessian = torch.load('./trained_weights/vggHessian_seed{}.pth'.format(s), map_location=torch.device('cpu'))
    
    mns = meancalc(Hessian)
    
    layermeans['zero_weight'].append(mns[0])
    layermeans['zero_bias'].append(mns[1])
    layermeans['one_weight'].append(mns[2])
    layermeans['one_bias'].append(mns[3])
    layermeans['three_weight'].append(mns[4])
    layermeans['three_bias'].append(mns[5])
    layermeans['six_weight'].append(mns[6])
    layermeans['six_bias'].append(mns[7])
    layermeans['eleven_weight'].append(mns[8])
    layermeans['eleven_bias'].append(mns[9])
    layermeans['thirteen_weight'].append(mns[10])
    layermeans['thirteen_bias'].append(mns[11])
    layermeans['sixteen_weight'].append(mns[12])
    layermeans['sixteen_bias'].append(mns[13])
    layermeans['eighteen_weight'].append(mns[14])
    layermeans['eighteen_bias'].append(mns[15])
    layermeans['twentythree_weight'].append(mns[16])
    layermeans['twentythree_bias'].append(mns[17])
    layermeans['twentysix_weight'].append(mns[18])
    layermeans['twentysix_bias'].append(mns[19])
    layermeans['twentyeight_weight'].append(mns[20])
    layermeans['twentyeight_bias'].append(mns[21])
    
avgDict = {}
for k, v in layermeans.items():
    avgDict[k] = round(sum(v) / float(len(v)), 4)
    if 'weight' in k:
        print(k, avgDict[k])

    
```

```python
#import all relevant packages
import torch, torchvision
from prettytable import PrettyTable

def meancalc(Hessian_diag_x):
    vals = []
    for entry in Hessian_diag_x:
        mean = torch.mean(entry)
        vals.append(mean.item())
    return vals

for dataset in ['MNIST', 'KMNIST', 'EMNIST', 'FashionMNIST']:
    rowvals = []
    x = PrettyTable()
    x.field_names = ['prior', '0.weight', '0.bias', '3.weight', '3.bias', '7.weight', '7.bias']
    for prior in ['1', '01', '001', '0001', '00001', '000001']:
        layermeans = {
            'one_weight': [],
            'one_bias': [],
            'three_weight': [],
            'three_bias': [],
            'seven_weight': [],
            'seven_bias': [],
        }
        for i in range(1, 11):
            s = i
            
            if dataset == 'MNIST':
                DATAPATH = './Hessians/{}/Hessian_prec{}_seed{}.pth'.format(dataset, prior, s)
            else:
                DATAPATH = './Hessians/{}/{}_Hessian_prec{}_seed{}.pth'.format(dataset, dataset, prior, s)

            XMNIST_NN_Std_prec_x = torch.load(DATAPATH, map_location=torch.device('cpu'))

            mns = meancalc(XMNIST_NN_Std_prec_x)

            layermeans['one_weight'].append(mns[0])
            layermeans['one_bias'].append(mns[1])
            layermeans['three_weight'].append(mns[2])
            layermeans['three_bias'].append(mns[3])
            layermeans['seven_weight'].append(mns[4])
            layermeans['seven_bias'].append(mns[5])

        avgDict = {}
        for k, v in layermeans.items():

            avgDict[k] = round(sum(v) / float(len(v)), 4)

        #print('avg on {} with prior {} is:'.format(dataset, prior), list(avgDict.values()))
        #print()
        dictvals = list(avgDict.values())
        dictvals.insert(0, prior)
        
        
        x.add_row(dictvals)
    print("average unvertainty in {}: ".format(dataset) )
    print(x)
    print()
    
```
